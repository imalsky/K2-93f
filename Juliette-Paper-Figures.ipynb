```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('/home/imalsky/isaac-style.txt')
```

```python
file = '/home/imalsky/Documents/New-Juliette/LOGS/evolve_12_0.75_0.24_0.02_1.37_7.48_0.1.mod'

df = pd.read_csv(file, delimiter=r"[^\S\r\n]{2,}", skiprows=5)

print (list(df))
```

```python
# Figure aesthetics
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 14), squeeze=False, sharex=True)
plt.subplots_adjust(hspace=0.15, wspace=0.23)


ax[0,0].plot(10 ** df.log_star_age[2:], df["R_transit (1d-3 bar, 1d3 barye)"][2:] / 637.1e6,
            linewidth=3, color='black')
ax[0,0].set_xscale('log')
ax[0,0].set_xlabel('Age (Years)', fontsize=24)
ax[0,0].set_ylabel(r'Transit Radius (M$_\oplus$)', fontsize=24)


ax[0,1].plot(10 ** df.log_star_age[2:], df["Temperature of base"][2:], linewidth=3, color='black')
ax[0,1].set_xscale('log')
ax[0,1].set_xlabel('Age (Years)', fontsize=24)
ax[0,1].set_ylabel(r'Temp at Atm Base (K)', fontsize=24)


ax[1,0].plot(10 ** df.log_star_age[2:-1], df["T(1)"][2:-1], linewidth=3, color='black')
ax[1,0].set_xscale('log')
ax[1,0].set_xlabel('Age (Years)', fontsize=24)
ax[1,0].set_ylabel(r'Temp at $\tau=2/3$ (K)', fontsize=24)


ax[1,1].plot(10 ** df.log_star_age[2:-1], df["Pressure of Base"][2:-1], linewidth=3, color='black')
ax[1,1].set_xscale('log')
ax[1,1].set_xlabel('Age (Years)', fontsize=24)
ax[1,1].set_ylabel(r'Pressure at Atm Base (dyne)', fontsize=24)

plt.savefig('/home/imalsky/Desktop/evolution-figures.png', dpi=200, bbox_inches='tight')
```
